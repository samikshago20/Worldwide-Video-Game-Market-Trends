```python
import pandas as pd
video_games=(
    pd.read_csv("/content/vgchartz-2024.csv",parse_dates=["release_date"])
    .rename({
        "title":"Title",
        "console":"Console",
        "genre":"Genre",
        "publisher":"Publisher",
        "developer":"Developer"
    },axis=1)
    .assign(release_year=lambda x:x["release_date"].dt.year)
   )
video_games.head()
```

# **Objective 2:**

```python
annual_sales=video_games.groupby("release_year").agg({"total_sales":sum})
annual_sales.tail()
```

```python
import plotly.express as px
annual_sales = annual_sales.reset_index()
px.line(annual_sales,x="release_year",y="total_sales")
```

```python
top10_titles=(
    video_games.groupby("Title",as_index=False)
    .agg({"total_sales":"sum"})
    .sort_values(by="total_sales",ascending=False)
    .iloc[:10]
)
top10_titles.head()
```

```python
import plotly.express as px

# Reset the index to make 'Title' a column
top10_titles = top10_titles.reset_index()

# Now you can use 'Title' as the x-axis
px.bar(top10_titles, x="Title", y="total_sales")
```
